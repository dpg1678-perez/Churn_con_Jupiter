## 7.1 DONDE HEMOS PERDIDO MAS CLIENTE
PARA LO CUAL SEGMENTAREMOS POR LOCALIDAD Y HAREMOS UN CONTEO
##7.2 DONDE HEMOS PERDIDO MAS CLIENTE EN PORCENTAJE
PARA LO CUAL COGEREMOS EL TOTAL Y HAREMOS EL %

```python
localidades = list(set(users["City"]))
#print(localidades)
localidadesordenadas = sorted(localidades)
#print(localidadesordenadas)
totalporlocalidad = dict()
totalabandonosporlocalidad = dict()
#print(users["Count_x"])
for localidad in localidadesordenadas:
    totalporlocalidad[localidad] = users[users['City'].isin([localidad])].loc[:,"Count_x"].sum()
    totalabandonosporlocalidad[localidad] = users[users['City'].isin([localidad]) & users['Churn Value'].isin([1])].loc[:,"Count_x"].sum()
    #print(users[users['City'].isin([localidad])].loc[:,["Customer ID", "City", "Count_x", "Churn Value"]])
print(totalporlocalidad)
print(totalabandonosporlocalidad)
maximoabandono = 0
listamaximoabandono=[]
maximoporcentaje = 0
listamaximoporcentaje=[]
for localidad in localidadesordenadas:
    if totalporlocalidad[localidad] > maximoabandono:
        maximoabandono = totalporlocalidad[localidad]
        listamaximoabandono=[]
    if totalporlocalidad[localidad] == maximoabandono:
        listamaximoabandono.append(localidad)
    if localidad in totalabandonosporlocalidad:
        if (100*totalabandonosporlocalidad[localidad] / totalporlocalidad[localidad]) > maximoporcentaje:
            listamaximoporcentaje=[]
            maximoporcentaje = 100*totalabandonosporlocalidad[localidad] / totalporlocalidad[localidad]
        if maximoporcentaje == (100*totalabandonosporlocalidad[localidad] / totalporlocalidad[localidad]):
            listamaximoporcentaje.append(localidad)
print(maximoabandono)
print(maximoporcentaje)
print(listamaximoabandono)
print(listamaximoporcentaje)
```

```python
# PARA HALLAR EL PORCENTAJE DE LOS ANGELES, SE QUE NO HAY MAS QUE 1 PERO ES POR SI HUBIERA MAS
for localidad in listamaximoabandono:
    maximoporcentaje = 100*totalabandonosporlocalidad[localidad] / totalporlocalidad[localidad]
    print(f"{localidad} tiene una perdida de {maximoporcentaje}")
# PARA ENCONTRAR EL NUMERO PERDIDO EB DONDE SE HA PERDIDO EL 100% DE CONTRATOS
for localidad in listamaximoporcentaje:
    print(f"{localidad} tiene perdidaas de {totalabandonosporlocalidad[localidad]}")
```

#### 7.5 ACEPTAR OFERTAS AFECTA EN EL ABANDONO


```python
tipooferta = list(set(users["Offer"]))
print(tipooferta)
tipooferta2 = tipooferta.copy()
for oferta in tipooferta:
    if pd.isna(oferta):  # CON ESTO VALE para quitar la nan
        print("capture NAN")
        tipooferta2.pop(tipooferta2.index(oferta))
print(tipooferta2)

# print(tipoofertasinnan)
ofertas = dict()
ofertastotales = 0
ofertasabandono = dict()
ofertastotalesabandono =0
for oferta in tipooferta2:
    # print(users[users["Offer"].isin([oferta])].loc[:,["Customer ID", "Offer", "Count_x"]])
    ofertas[oferta] = users[users["Offer"].isin([oferta])].loc[:,"Count_x"].sum()
    ofertastotales += ofertas[oferta]
    #print(users["Churn Value"])
    #print(users[users["Churn Value"]==1 & users["Offer"].isin([oferta])].loc[:,"Count_x"].sum())
    ofertasabandono[oferta] = users[users["Offer"].isin([oferta]) & users["Churn Value"]==1].loc[:,"Count_x"].sum()
    ofertastotalesabandono += ofertasabandono[oferta]

print(ofertas)
print(ofertasabandono)
print(ofertastotales)
print(ofertastotalesabandono)
# hablamos de cantidad

listaabandonanofertavalor = list(map(int, ofertasabandono.values()))
#print(listaabandonanofertavalor)
listaabandonanofertavalorord = sorted(listaabandonanofertavalor)
print(listaabandonanofertavalorord)
for valor in listaabandonanofertavalorord:
      for causa in tipooferta2:
          if ofertasabandono[causa] == valor:
              print(f"El puesto {5-listaabandonanofertavalorord.index(valor)} es para {causa} con {ofertasabandono[causa]} bajas ")
# hablamos de porcentaje
              print(f"Esto representa el {ofertasabandono[causa]*100/ofertastotalesabandono:.2f}")
```
